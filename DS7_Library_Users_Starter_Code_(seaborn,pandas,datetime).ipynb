```
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
```

```
# STARTER CODE

# List of Lists Method
# personID1, personFirstName1, personLastName1, and dateJoined1 represent columns. They all have the same length
# zip() takes the first element in each list and creates a new list. 
# Then it takes the second element in each list and creates a new list. It does this for every element in the list.
# list() converts the result of zip() into a list of lists A.K.A a list of rows for the new DataFrame  

personID1 = ["0001", "0002", "0003", "0004", "0005", "0006", "0007", "0008", "0009", "0010"]
personFirstName1 = ["Albert", "Jackie", "Jasper", "Tina", "Eliza", "Sarah", "Alexander", "Robert", "Shaun", "Cally"]
personLastName1 = ["Mobley", "Comstock", "Bosma", "Canessa", "Knuth", "Tate", "Martel", "Kravitz", "Schopf","Colins"]
dateJoined1 = ["12/02/2007", "12/02/2007","12/02/2007", "12/03/2007", "12/05/2007", "12/05/2007", "12/06/2007", "12/06/2007", "12/06/2007", "12/10/2007"]

# 1) PersonDF1
# 2) DateDF1
personDF1 = pd.DataFrame(list(zip(personID1,personFirstName1,personLastName1)), columns=["ID","First Name", "Last Name"])
dateDF1 = pd.DataFrame(list(zip(personID1,dateJoined1)), columns = ["ID", "Date Joined"])
dateDF1["Date Joined"] = pd.to_datetime(dateDF1["Date Joined"])


# Dictionary Method
# personID2, personFirstName2, personLastName2, and dateJoined2 represent columns. They all have the same length
# We create a dictionary where the keys are the column labels and the values are the columns
# This works the same way as the list of lists method but it's simpler

personID2 = ["0101", "0102", "0103", "0104", "0105", "0106", "0107", "0108", "0109", "0110"]
personFirstName2 = ["Madison", "Cameron", "Ephraim", "Elizabeth", "Charlotte", "Sophie", "Luke", "Michael", "Babe", "Janet"]
personLastName2 = ["Keith", "Bate", "Willie", "Vera", "Vergara", "Mitchell", "Lewis", "Bliss", "Opie","Kim"]
dateJoined2 = ["02/22/2008", "02/22/2008","02/25/2008", "02/27/2008", "03/02/2008", "03/05/2008", "03/06/2008", "03/10/2008", "03/13/2008", "03/13/2008"]

# 3) PersonDF2
# 4) DateDF2
personDF2 = pd.DataFrame({"ID":personID2,"First Name":personFirstName2,"Last Name":personLastName2})
dateDF2 = pd.DataFrame({"ID":personID2,"Date Joined":dateJoined2})
dateDF2["Date Joined"] = pd.to_datetime(dateDF2["Date Joined"])

favoriteBook = ["1984", "Twilight", "Harry Potter", "To Kill a Mocking Bird", "The Lion, the Witch and the Wardrobe",  "Catch-22", "Harry Potter", "Gone with the Wind", "Animal Farm", "Possession", "1984", "The Stormlight Archives", "It", "The Hunger Games", "Eragon", "The Lion, the Witch, and the Wardrobe", "Divergent", "If Only to Deceive", "Dune", "1984"]
LibaryCardExpires = ["06/01/2016", "08/01/2016","08/01/2016", "09/01/2016", "05/01/2016", "04/01/2016", "03/01/2017", "04/01/2016", "02/01/2017", "03/01/2017","05/01/2016", "11/01/2016","08/01/2016", "12/01/2016", "12/01/2016", "10/01/2016", "01/01/2017", "01/01/2017", "02/01/2017", "07/01/2016"]

# 5) BooksCardDF
BooksCardDF = pd.DataFrame({"Favorite Book":favoriteBook,"Library Card Expiration Date":LibaryCardExpires})
BooksCardDF["Library Card Expiration Date"] = pd.to_datetime(BooksCardDF["Library Card Expiration Date"])

# Print all the DataFrames:
print("Person Table #1")
display(personDF1)
print("\nJoin Date Table #1")
display(dateDF1)
print("\nPerson Table #2")
display(personDF2)
print("\nJoin Date Table #2")
display(dateDF2)
print("\nFavorite Book and Card Expiration Table #2")
display(BooksCardDF)
```

```
combineDF1 = pd.merge(personDF1,dateDF1, on = "ID")
display(combineDF1)
```

```
combineDF2 = pd.merge(personDF2,dateDF2, on = "ID")
display(combineDF2)
```

```
#Concatenate rows
combinedRows = pd.concat([combineDF1,combineDF2],ignore_index=True)

#Reset index can also be done as below
#combinedRows.reset_index(drop=True,inplace=True)

display(combinedRows)
```

```
combinedCols = pd.concat([combinedRows,BooksCardDF],axis=1)

display(combinedCols)
```
